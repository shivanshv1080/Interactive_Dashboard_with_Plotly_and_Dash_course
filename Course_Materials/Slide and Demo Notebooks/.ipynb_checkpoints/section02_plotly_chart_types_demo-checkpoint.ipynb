```python
import pandas as pd
import numpy as np

import plotly.graph_objects as go
import plotly.express as px
```

```python
education = pd.read_csv("states_all.csv").assign(
    expenditure_per_student = lambda x: x["TOTAL_EXPENDITURE"] / x["GRADES_ALL_G"],
    above_avg_math8 = lambda x: np.where(x["AVG_MATH_8_SCORE"] > x["AVG_MATH_8_SCORE"].mean(),'Above Avg', 'Below Avg')
)

```

# Bar Charts

```python
px.line(
    education,
    x="YEAR",
    y="FEDERAL_REVENUE"
)
```

```python
px.line(
    education.query("STATE == 'CALIFORNIA'"),
    x="YEAR",
    y="FEDERAL_REVENUE",
    title="California Annual Federal Revenue"
)
```

```python
px.line(
    education.query("STATE in ['CALIFORNIA', 'TEXAS']"),
    x="YEAR",
    y="FEDERAL_REVENUE",
    markers=True,
    color='STATE',
    title="California Annual Federal Revenue"
)
```

```python
px.line(
    education.query("STATE in ['CALIFORNIA', 'TEXAS']"),
    x="YEAR",
    y="FEDERAL_REVENUE",
    markers=True,
    color='STATE',
    symbol='STATE',
    title="California Annual Federal Revenue"
)
```

# Scatterplots

```python
px.scatter(
    education, 
    x="FEDERAL_REVENUE", 
    y="STATE_REVENUE",
    trendline="ols",
    marginal_x="box",
    marginal_y="box",
    color="STATE", # can be continous
    size="YEAR", # can't be missing
    title = "State Vs. Federal Revenue"
)
```

```python
px.scatter(education, 
           x="FEDERAL_REVENUE", 
           y="STATE_REVENUE",
           color="STATE"
          )
```

```python
import plotly.express as px
df = px.data.gapminder().query("year == 2007").query("continent == 'Americas'")
fig = px.pie(df, values='pop', names='country',
             title='Population of American continent',
             hover_data=['lifeExp'], labels={'lifeExp':'life expectancy'})
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```
